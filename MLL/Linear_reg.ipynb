```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data from train.csv | test.csv inside Datum directory
train = pd.read_csv('Datum/train.csv')
test = pd.read_csv('Datum/test.csv')


plt.plot(train['x'], train['y'], 'o', label='Original data')
plt.plot(test.x, test.y, 'x', label='Testing data')
plt.legend()
```

```python
X_train, y_train = train['x'].values.reshape(-1,1), train['y'].values.reshape(-1,1)
X_test, y_test = test['x'].values.reshape(-1,1), test['y'].values.reshape(-1,1) # reshape(-1,1) is used to convert 1D array to 2D array
for i in range(len(y_train)):
    try:
        int(y_train[i])
    except:
        y_train[i] = y_train[i - 1]
for i in range(1,len(X_train)):
#     if X_train[i][0] > 80:
#         X_train[i][0] = 50
    if X_train[i][0] == X_train.max():
        X_train[i][0] = 50# X_train[i-1][0]
        print("one goes")
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
# Weight and bias
print('Weight: ', lr.coef_)
print('Bias: ', lr.intercept_)
```

```python
# Accuracy
print('Training accuracy: ', lr.score(X_train, y_train))
print('Testing accuracy: ', lr.score(X_test, y_test))
```

```python
# Plotting the Model and Data
plt.plot(X_train, y_train, '.', label='Original data')
plt.plot(X_test, y_test, '.', label='Testing data')
plt.plot(X_train, lr.predict(X_train), "s", label='Fitted line')
plt.legend()
```
